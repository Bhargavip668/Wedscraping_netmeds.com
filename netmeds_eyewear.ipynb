```python
productname = []
pricebox = []
marketed = []
savetag = []
pagenum = []


for i in range(1,20):
    start_time = time.time()
    URL = ' https://www.netmeds.com/non-prescriptions/eyewear/page/'+str(i)
    page = requests.get(URL)
    pagecontent = page.text
    soup = BeautifulSoup(pagecontent)
    
    for x in soup.find_all('div' , attrs = {'class' : 'col-md-3 padding-none'}):
        pname = x.find('span' , attrs = {'class' : 'clsgetname'})
        price = x.find('span' , attrs = {'class' : 'price-box'})
        mkt = x.find('span' , attrs = {'class' : 'drug-varients ellipsis'})
        save = x.find('span' , attrs = {'class' : 'save-badge'})
        if pname is None:
            productname.append(np.NaN)
        else:
            productname.append(pname.text)
        if price is None:
            pricebox.append(np.NaN)
        else:
            pricebox.append(price.text)
        if mkt is None:
            marketed.append(np.NaN)
        else:
            marketed.append(mkt.text)
        if save is None:
            savetag.append(np.NaN)
        else:
            savetag.append(save.text)
        pagenum.append(i)
    print('page {} completed in {} seconds'.format(i,time.time()-start_time))        
```

```python
len(productname)
```

```python
laptop_df = pd.DataFrame({'Product' : productname , 'PriceBox' : pricebox , 'MarketedBy' : marketed , 'Savings' : savetag })
```

```python
laptop_df.head()
```

```python
laptop_df.to_csv('netmeds_eyewear.csv')
```

```python
laptop_df.shape
```

```python
regex = 'Rs.\s(.*)\sMRP'
laptop_df['BestPrice'] = laptop_df['PriceBox'].apply(lambda x : re.compile(regex).findall(x))
```

```python
laptop_df.shape
```

```python
laptop_df.head()
```

```python
regex = 'MRP.Rs.(.*)'
laptop_df['MRPPrice'] = laptop_df['PriceBox'].apply(lambda x : re.compile(regex).findall(x))
```

```python
laptop_df.head()
```

```python
laptop_df['BestPrice'] = laptop_df['BestPrice'].apply(lambda x : ''.join(x))
laptop_df['MRPPrice'] = laptop_df['MRPPrice'].apply(lambda x : ''.join(x))
```

```python
laptop_df.head()
```

```python
laptop_df['BestPrice'] = pd.to_numeric(laptop_df['BestPrice'],errors='coerce')
laptop_df['MRPPrice'] = pd.to_numeric(laptop_df['MRPPrice'],errors='coerce')
```
